```python
def calculate_iqr(numbers):
    q75, q25 = np.percentile(numbers, [75, 25])
    return q75 - q25
```

```python
def calculate_coefficient_of_variation(numbers):
    mean = np.mean(numbers)
    std_dev = np.std(numbers, ddof=1)
    return std_dev / mean
```

```python
def calculate_mad(numbers):
    mean = np.mean(numbers)
    return np.mean(np.abs(numbers - mean))
```

```python
def calculate_quartile_deviation(numbers):
    q75, q25 = np.percentile(numbers, [75, 25])
    return (q75 - q25) / 2
```

```python
def calculate_range_based_coefficient_of_dispersion(numbers):
    range_value = np.max(numbers) - np.min(numbers)
    mean = np.mean(numbers)
    return range_value / mean
```

# Q3) Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

```python
import numpy as np

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not np.isclose(sum(probabilities), 1):
            raise ValueError("Probabilities must sum to 1.")
        
        self.values = np.array(values)
        self.probabilities = np.array(probabilities)

    def expected_value(self):
        return np.sum(self.values * self.probabilities)

    def variance(self):
        mean = self.expected_value()
        return np.sum((self.values - mean) ** 2 * self.probabilities)

# Example usage
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

drv = DiscreteRandomVariable(values, probabilities)

expected_value = drv.expected_value()
variance = drv.variance()

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
```

# 4) Implement a program to simulate the rolling o6 a 6air six-sided die and calculate the expected value and variance o6 the outcomes.

```python
import numpy as np

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not np.isclose(sum(probabilities), 1):
            raise ValueError("Probabilities must sum to 1.")
        
        self.values = np.array(values)
        self.probabilities = np.array(probabilities)

    def expected_value(self):
        return np.sum(self.values * self.probabilities)

    def variance(self):
        mean = self.expected_value()
        return np.sum((self.values - mean) ** 2 * self.probabilities)

def simulate_dice_rolls():
    outcomes = [i + j for i in range(1, 7) for j in range(1, 7)]
    unique_outcomes, counts = np.unique(outcomes, return_counts=True)
    probabilities = counts / sum(counts)
    
    return unique_outcomes, probabilities

# Simulate dice rolls
values, probabilities = simulate_dice_rolls()

# Create a DiscreteRandomVariable instance for the dice rolls
dice_rolls = DiscreteRandomVariable(values, probabilities)

# Calculate expected value and variance
expected_value = dice_rolls.expected_value()
variance = dice_rolls.variance()

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
```

#  Create a Pthon 6unction to generate random samples 6rom a given probabilit distribution (e.g., binomial, Poisson) and calculate their mean and variance.

```python


def generate_samples_and_calculate_stats(distribution, params, num_samples):
    """
    Generate random samples from a specified distribution and calculate their mean and variance.
    
    Parameters:
    distribution (str): The type of distribution ('binomial', 'poisson', etc.).
    params (dict): The parameters for the specified distribution.
    num_samples (int): The number of samples to generate.
    
    Returns:
    tuple: A tuple containing the mean and variance of the generated samples.
    """
    if distribution == 'binomial':
        n = params.get('n', 1)
        p = params.get('p', 0.5)
        samples = np.random.binomial(n, p, num_samples)
    elif distribution == 'poisson':
        lam = params.get('lam', 1)
        samples = np.random.poisson(lam, num_samples)
    else:
        raise ValueError("Unsupported distribution type. Supported types: 'binomial', 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples, ddof=1)
    
    return mean, variance

# Example usage
num_samples = 1000

# Binomial distribution example
binomial_params = {'n': 10, 'p': 0.5}
binomial_mean, binomial_variance = generate_samples_and_calculate_stats('binomial', binomial_params, num_samples)
print(f"Binomial Distribution - Mean: {binomial_mean}, Variance: {binomial_variance}")

# Poisson distribution example
poisson_params = {'lam': 4}
poisson_mean, poisson_variance = generate_samples_and_calculate_stats('poisson', poisson_params, num_samples)
print(f"Poisson Distribution - Mean: {poisson_mean}, Variance: {poisson_variance}")
```

# Write a Pthon script to generate random numbers 6rom a Gaussian (normal) distribution and compute the mean, variance, and standard deviation o6 the samples.

```python
import numpy as np

def generate_gaussian_samples(mean, std_dev, num_samples):
    samples = np.random.normal(mean, std_dev, num_samples)
    calculated_mean = np.mean(samples)
    calculated_variance = np.var(samples, ddof=1)
    calculated_std_dev = np.std(samples, ddof=1)
    
    return calculated_mean, calculated_variance, calculated_std_dev

# Example usage
mean = 0
std_dev = 1
num_samples = 1000

calculated_mean, calculated_variance, calculated_std_dev = generate_gaussian_samples(mean, std_dev, num_samples)

print(f"Generated Gaussian Samples - Mean: {calculated_mean}")
print(f"Generated Gaussian Samples - Variance: {calculated_variance}")
print(f"Generated Gaussian Samples - Standard Deviation: {calculated_std_dev}")
```

# Write a Pthon 6unction to calculate the probabilit densit 6unction (PDF) o6 a continuous random variable 6or a given normal distribution.


```python
import numpy as np

def calculate_normal_pdf(x, mean, std_dev):
    return (1 / (np.sqrt(2 * np.pi) * std_dev)) * np.exp(-(x - mean)**2 / (2 * std_dev**2))

# Example usage
x_values = np.linspace(-5, 5, 100)
mean = 0
std_dev = 1
pdf_values = calculate_normal_pdf(x_values, mean, std_dev)

# Print the calculated PDF values
for x, pdf in zip(x_values, pdf_values):
    print(f"x = {x:.2f}, PDF = {pdf:.5f}")
```

# ¸) Create a program to calculate the cumulative distribution 6unction (CDF) o6 exponential distribution.

```python
import numpy as np

def exponential_cdf(x, rate):
    return 1 - np.exp(-rate * x)

# Example usage
x_values = np.linspace(0, 5, 100)
rate = 0.5
cdf_values = exponential_cdf(x_values, rate)

# Print the calculated CDF values
for x, cdf in zip(x_values, cdf_values):
    print(f"x = {x:.2f}, CDF = {cdf:.5f}")
```

# Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

```python
import numpy as np
import math

def poisson_pmf(k, rate):
    
    if isinstance(k, int):
        return np.exp(-rate) * (rate**k) / math.factorial(k)
    else:
        return np.exp(-rate) * (rate**k) / np.array([math.factorial(ki) for ki in k])

# Example usage
k_values = np.arange(0, 11)  # Values of k from 0 to 10
rate = 2.5
pmf_values = poisson_pmf(k_values, rate)

# Print the calculated PMF values
for k, pmf in zip(k_values, pmf_values):
    print(f"k = {k}, PMF = {pmf:.5f}")
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Data for the old layout
old_layout = np.array([1] * 50 + [0] * 950)

# Data for the new layout
new_layout = np.array([1] * 70 + [0] * 930)

# Number of successes (purchases) and total observations (visitors) for each layout
successes = np.array([np.sum(old_layout), np.sum(new_layout)])
observations = np.array([len(old_layout), len(new_layout)])

# Perform z-test for proportions
z_stat, p_value = proportions_ztest(successes, observations, alternative='larger')

# Print the results
print("Z-test Statistic:", z_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is significant evidence to reject the null hypothesis. The new layout leads to a higher conversion rate.")
else:
    print("There is no significant evidence to reject the null hypothesis. The new layout does not lead to a higher conversion rate.")
```

```python
import numpy as np
from scipy.stats import norm

# Data for scores before and after the program
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate the differences in scores before and after the program
differences = after_program - before_program

# Calculate the mean and standard deviation of the differences
mean_difference = np.mean(differences)
std_dev_difference = np.std(differences, ddof=1)  # ddof=1 for sample standard deviation

# Calculate the standard error of the mean difference
standard_error = std_dev_difference / np.sqrt(len(differences))

# Calculate the z-score
z_score = mean_difference / standard_error

# Calculate the p-value
p_value = 2 * norm.cdf(-np.abs(z_score))  # Multiply by 2 for two-tailed test

# Print the results
print("Mean Difference:", mean_difference)
print("Standard Error of the Mean Difference:", standard_error)
print("Z-score:", z_score)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There is significant evidence to reject the null hypothesis.")
    print("The tutoring program improves students' exam scores.")
else:
    print("There is no significant evidence to reject the null hypothesis.")
    print("The tutoring program does not improve students' exam scores.")
```

```python
import numpy as np
from scipy.stats import ttest_ind

def ab_test(layout_a_clicks, layout_b_clicks):
    # Calculate the t-statistic and p-value using two-sample t-test
    t_statistic, p_value = ttest_ind(layout_a_clicks, layout_b_clicks)
    
    # Calculate degrees of freedom
    dof = len(layout_a_clicks) + len(layout_b_clicks) - 2
    
    return t_statistic, p_value, dof

# Data for click-through rates for layout A and layout B
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test analysis
t_statistic, p_value, dof = ab_test(layout_a_clicks, layout_b_clicks)

# Print the results
print("t-statistic:", t_statistic)
print("p-value:", p_value)
print("Degrees of Freedom:", dof)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There is significant evidence to reject the null hypothesis.")
    print("There is a significant difference between the click-through rates of the two layouts.")
else:
    print("There is no significant evidence to reject the null hypothesis.")
    print("There is no significant difference between the click-through rates of the two layouts.")
```

```python
import numpy as np
from scipy.stats import ttest_rel

def intervention_effect(pre_intervention_scores, post_intervention_scores):
    # Calculate the t-statistic and p-value using paired-sample t-test
    t_statistic, p_value = ttest_rel(pre_intervention_scores, post_intervention_scores)
    
    # Calculate degrees of freedom
    dof = len(pre_intervention_scores) - 1
    
    return t_statistic, p_value, dof

# Data for pre-intervention and post-intervention test scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Analyze the impact of the intervention program
t_statistic, p_value, dof = intervention_effect(pre_intervention_scores, post_intervention_scores)

# Print the results
print("t-statistic:", t_statistic)
print("p-value:", p_value)
print("Degrees of Freedom:", dof)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There is significant evidence to reject the null hypothesis.")
    print("The intervention had a significant impact on math scores.")
else:
    print("There is no significant evidence to reject the null hypothesis.")
    print("The intervention did not have a significant impact on math scores.")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

np.random.seed(0)

# Generate data
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create contingency table
contingency_table = np.zeros((3, 2), dtype=int)

for i, age_group in enumerate(['18-30', '31-50', '51+']):
    for j, preference in enumerate(['Candidate A', 'Candidate B']):
        contingency_table[i, j] = np.sum((age_groups == age_group) & (voter_preferences == preference))

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Statistic:", chi2_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Contingency Table:")
print(contingency_table)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There is significant evidence to reject the null hypothesis.")
    print("There is a significant association between age groups and voter preferences.")
else:
    print("There is no significant evidence to reject the null hypothesis.")
    print("There is no significant association between age groups and voter preferences.")
```
